```python
#Import Libraries
import pandas as pd
import numpy as np # linear algebra
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
file_path = "/home/marshal/protonotebook/Management and analysis of Biological and Medical Data/updated_file_tfs.csv"
file_path_2 = "/home/marshal/protonotebook/Management and analysis of Biological and Medical Data/updated_file_enzymes.csv"

#Read Tables
tfs = pd.read_csv (file_path)
enzymes = pd.read_csv(file_path_2)
```

### Box Plots


```python
# Select numeric columns only
numeric_cols = enzymes.select_dtypes(include='number').columns

# Set up the figure
plt.figure(figsize=(16, 8))

# Iterate over the numeric columns to create box plots
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, len(numeric_cols)//2 + 1, i)  # Create subplots for each column
    sns.boxplot(y=enzymes[col], palette='Set2')
    plt.title(f'Box plot of {col}')
    plt.tight_layout()

plt.show()
```
