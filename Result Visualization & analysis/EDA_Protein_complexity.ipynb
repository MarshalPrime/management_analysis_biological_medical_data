## Protein Complexity EDA

```python
#Import Libraries
import pandas as pd
import numpy as np # linear algebra
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
file_path = "/home/marshal/protonotebook/Management and analysis of Biological and Medical Data/updated_file_tfs.csv"
file_path_2 = "/home/marshal/protonotebook/Management and analysis of Biological and Medical Data/updated_file_enzymes.csv"

#Read Tables
tfs = pd.read_csv (file_path)
enzymes = pd.read_csv(file_path_2)
```

## Box Plots


### Enzymes 

```python
# Select numeric columns only
numeric_cols = enzymes.select_dtypes(include='number').columns

# Set up the figure
plt.figure(figsize=(16, 8))

# Iterate over the numeric columns to create box plots
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, len(numeric_cols)//2 + 1, i)  # Create subplots for each column
    sns.boxplot(y=enzymes[col], palette='Set2')
    plt.title(f'Box plot of {col}')
    plt.tight_layout()

plt.show()
```

## Compare Results Eznymes & Tfs By (Shannon Entropy,Number of Transitions,Fractal Dimension)

### Shannon Entropy

```python
# Create a figure and axis with subplots
plt.figure(figsize=(12, 6))

# Box plot for Enzymes
plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
sns.boxplot(y=enzymes['Shannon Entropy'], color='lightblue')
plt.title('Enzymes Complexity')
plt.ylabel('Shannon Entropy')

# Box plot for TFs
plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
sns.boxplot(y=tfs['Shannon Entropy'], color='lightgreen')
plt.title('TFs Complexity')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

### Number of Transitions

```python
# Create a figure and axis with subplots
plt.figure(figsize=(12, 6))

# Box plot for Enzymes
plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
sns.boxplot(y=enzymes['Number of Transitions'], color='lightblue')
plt.title('Enzymes Complexity')
plt.ylabel('Number of Transitions')

# Box plot for TFs
plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
sns.boxplot(y=tfs['Number of Transitions'], color='lightgreen')
plt.title('TFs Complexity')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

### Fractal Dimension

```python
# Create a figure and axis with subplots
plt.figure(figsize=(12, 6))

# Box plot for Enzymes
plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
sns.boxplot(y=enzymes['Fractal Dimension'], color='lightblue')
plt.title('Enzymes Complexity')
plt.ylabel('Fractal Dimension')

# Box plot for TFs
plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
sns.boxplot(y=tfs['Fractal Dimension'], color='lightgreen')
plt.title('TFs Complexity')
plt.ylabel('')

plt.tight_layout()
plt.show()
```
